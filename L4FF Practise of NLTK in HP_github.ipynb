```python
global words_count,words_count_removing_punctuation,words_count_unique_removing_punctuation
global lexical_diversity_clean,lexical_diversity_punctuation,avg_sent_length,words_count_clean
words_count=[]
words_count_clean=[]
words_count_removing_punctuation=[]
words_count_unique_removing_punctuation=[]
lexical_diversity_clean=[]
lexical_diversity_punctuation=[]
avg_sent_length=[]
```

```python
import nltk
import nltk.draw.dispersion
def extract(txt):
    with open(txt,encoding="utf-8") as f:
        content=f.read().lower()
        
    num_word=nltk.word_tokenize(content) 
    print(f'There are {len(num_word)} words before cleansing')
    words_count.append(len(num_word))
    
    num_sent=nltk.sent_tokenize(content)
    print(f'There are {len(num_sent)} sentences before cleansing')

    from nltk.corpus import stopwords
    from string import punctuation
    custom_stopwords= set(stopwords.words('english')+list(punctuation))
    word_text_clean=[word for word in num_word if word not in custom_stopwords]
    words_count_clean.append(len(word_text_clean))
    
    num_words_clean=len(word_text_clean)
    print(f'There are {num_words_clean} words after cleansing')
    
    word_text_clean_stop=[word for word in nltk.word_tokenize(content) if word not in punctuation]
    
    num_words_clean_stop=len(word_text_clean_stop)
    print(f'There are {num_words_clean_stop} words after only removing punctuation')
    words_count_removing_punctuation.append(num_words_clean_stop)

    
    unique_word_text_stop=set(word_text_clean_stop)
    print(f'There are {len(unique_word_text_stop)} unique words after only removing punctuation')
    words_count_unique_removing_punctuation.append(len(unique_word_text_stop))

    unique_word_text=set(word_text_clean)
    print(f'There are {len(unique_word_text)} unique words after cleansing')

    lexical_diversity= len(unique_word_text_stop)/num_words_clean_stop
    print(f'lexical_diversity is {lexical_diversity} after only removing punctuation')
    lexical_diversity_punctuation.append(lexical_diversity)
    
    print(f'lexical_diversity is {len(unique_word_text)/num_words_clean} after cleansing')
    lexical_diversity_clean.append(len(unique_word_text)/num_words_clean)

    print(f' Average sentence length is {num_words_clean_stop/len(num_sent)} after cleansing')
    avg_sent_length.append(num_words_clean_stop/len(num_sent))

    
    from nltk.text import Text
    HP1=Text(word_text_clean)
    
    import matplotlib.pyplot as plt
    plt.rcParams.update({'font.size': 22})

    plt.figure(figsize=(20, 12))
    
    nltk.draw.dispersion.dispersion_plot(text=HP1,words=['harry','ron','hermione','dumbledore','snape','voldemort','you-know-who','hagrid','professor','wand'],title=f'{txt_name}')
 
```
